```python
import numpy as np #linear algerbra
import pandas as pd# data processing, csv file
```

```python
import seaborn as sns # seaborn is a python data visulization librabry based on matplotlib
import matplotlib.pyplot as plt #Matlab is a low level graphn plotting librabrary
import plotly.express as px  # allow you u interactive plots with very little code
```

```python
# READ THE DATA FROM THE CSV FILES

# mental-and-substance-use-as-share-of-disease.csv
df1 = pd.read_csv('mental-and-substance-use-as-share-of-disease -AI.csv')
# prevalence-by-mental-and-substance-use-disorder.csv
df2 = pd.read_csv('prevalence-by-mental-and-substance-use-disorder _AI.csv')
```

```python
df1.head()
```

```python
df2.head(10)# get top 10 values
```

```python
# Merging Above Two data set
data = pd.merge(df1,df2)
data.head(10)
```

```python
#DATA CLEANING
# Detect Missing values in the dataset
data.isnull().sum()
```

```python
# Drop the column
data.drop('Code',axis=1,inplace=True)
```

```python
data.head(10)
```

```python
# size = row*column , shape= tuple of array dimension(row,column)
data.size,data.shape
```

```python
# Column set

data=data.set_axis(['Country', 'Year', 'Schizophrenia', 'Bipolar_disorder', 'Eating_disorders', 'Anxiety', 'drug_usage', 'depression', 'alcohol', 'mental_fitness'], axis='columns')
data.head(10)
```

```python
data.head(10)
```

```python
# Visualization
numeric_columns = data.select_dtypes(include=np.number)
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='Blues')
plt.plot()
```
